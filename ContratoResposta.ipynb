```python
import datetime
class Profissional:
    
    def __init__(self, codigo, nome, endereco, nascimento, experiencia):
        self._codigo = codigo
        self._nome = nome
        self._endereco = endereco
        self._datanasc = nascimento
        self._anosexp = experiencia
        self._profissoes = []
        self._profissionais_disponiveis = []
        
    def cadastrarProfissao(self, profissao):
        self._profissoes.append(profissao)
        
    @property
    def getProfissoes(self):
        return self._profissoes
    
    @property
    def getNome(self):
        return self._nome
      
    @property
    def getAnosexp(self):
        return self._anosexp
      
    def calcularIdade(self):
        d1 = datetime . date . today ()
        d2 = self._datanasc
        quantidade_dias = abs((d2 - d1).days)
        quantidade_dias = quantidade_dias // 365
        return quantidade_dias
      
    
    def cadastrarProfissionaisDisponiveis(self, Profissional):
        self._profissionais_disponiveis.append(Profissional)
        
    @property
    def getProfissionaisDisponiveis(self):
        return self._profissionais_disponiveis
        
    def printProfissionaisDisponiveis(self):
        #for i in range (len(self._profissionais_disponiveis)):
        #    print(self._profissionais_disponiveis[i])
        #for (i,p) in enumerate(self._profissionais_disponiveis):
        #    print(f'indice={i}, profissional={p}')
        for p in self._profissionais_disponiveis:
            print(p)
        
    def melhorQue(self, profissao, profissional):
        apt1 = False
        apt2 = False
        for profissao1 in self._profissoes:
            if profissao1 == profissao:
                apt1 = True
                break
        
        for profissao2 in profissional.getProfissoes:
            if profissao2 == profissao:
                apt2 = True
                break
        
        if apt1 == apt2:
            if( self._anosexp > profissional._anosexp ):
                return True
            else:
                return False
        elif apt1 == True:
            return True
        else:
            return False
        
```

```python
import datetime


class Profissao:
    
    def __init__(self):
        self._habilidades = []
       # self._profissionais = []
        
    def cadastrarHabilidade(self, habilidade):
        self._habilidades.append(habilidade)
        
    @property
    def getHabilidades(self):
        return self._habilidades
        
    def cadastrarProfissional(self, profissional):
        self._profissionais.append(profissional)
        
    def __str__(self):
        return str(self._habilidades)
    
    def __eq__(self, outro):
        isequal = False
        if len(self._habilidades) == len(outro.getHabilidades):
            isequal = True
            for index in range(len(self._habilidades)):
                 if not (self._habilidades[index] in outro.getHabilidades):
                    isequal = False
                    break                   
        return isequal    
                
class Contrato:
    
    def __init__(self, numero, vigencia, salario, profissional, empresa):
        self._numero = numero
        self._vigencia = vigencia ##datafim-datainicio
        self._salario = salario
        self._profissional = profissional
        self._empresa = empresa
        
class Oferta:
    
    def __init__(self, id, salario, profissao):
        self._id = id
        self._salario = salario
        self._profissao = profissao
        
    
    def profissionaisAptos(self, profissionais):
        profissionais_aptos = []
        ## p = profissional
        for p in profissionais: 
            for profissao in p.getProfissoes:
                
                if self._profissao == profissao:
                    profissionais_aptos.append(p)
                    break
        return profissionais_aptos
      
    def melhorCandidato(self, profissionais):
        if len(profissionais) > 0 :
            melhor_can = profissionais[0]
            for p in profissionais:
                if not melhor_can.melhorQue(self._profissao, p):
                    melhor_can = p
            return melhor_can
        else:
            return None
        
                   
        
class Empresa:
    
    def __init__(self, CNPJ, nome, endereco):
        self._CNPJ = CNPJ
        self._nome = nome
        self._endereco = endereco
        self._oferta = []
        
    def cadastraOferta(self, id, salario, profissao):
        oferta = Oferta(id, salario, profissao)
        self._oferta.append(oferta)
        print('Oferta {0} cadastrada'.format(id))
    
    @property
    def getOfertas(self):
        return self._oferta
        
```
